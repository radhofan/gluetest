### This Reproducible Paper Project can be ran either in trovi terminal or jupyter notebook bash in reproduce.ipynb
### If you decide to run with the notebook, then dont forget to change the kernel to bash instead of python in the top right corner

#### Deactivate default conda env from trovi

```bash
conda init
conda deactivate
```

#### Install zip

```bash
sudo apt update
sudo apt install zip
```

#### Install sdkman

```bash
curl -s "https://get.sdkman.io" | bash
source "$HOME/.sdkman/bin/sdkman-init.sh"
```

#### Initialize sdkman, change [user email] with your chameleon email, sdkman will tell the correct path

```bash
source "/home/radhofanazizi_gmail_com/.sdkman/bin/sdkman-init.sh"
```

#### Install java

```bash
sdk install java 17.0.7-graal
```

#### Install gu python

```bash
gu install python
```

#### Install maven

```bash
sdk install maven
```

#### Create mamba env

```bash
(mamba create --name gluetest python=3.11.4 -y && mamba init && source ~/.bashrc && mamba activate gluetest)
```

#### Install pytest 

```bash
mamba install pytest -y
mamba install hamcrest -y
```

#### Run all test

```bash
bash run.sh
```
